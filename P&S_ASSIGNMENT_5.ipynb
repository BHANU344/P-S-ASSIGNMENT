```python
#Q1
import numpy as np
import matplotlib.pyplot as plt

def question_1(N):
    calls_per_day = np.random.randint(0, 8, size=N)
    pmf_values = [np.sum(calls_per_day == i) / N for i in range(8)]
    cdf_values = np.cumsum(pmf_values)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    axes[0].stem(range(8), pmf_values, basefmt=" ")
    axes[0].set_title("PMF (Probability Mass Function)")
    axes[0].set_xlabel("Number of Calls")
    axes[0].set_ylabel("Probability")

    axes[1].step(range(8), cdf_values, where="post")
    axes[1].set_title("CDF (Cumulative Distribution Function)")
    axes[1].set_xlabel("Number of Calls")
    axes[1].set_ylabel("Cumulative Probability")

    plt.tight_layout()
    plt.show()

N_calls = 1000
question_1(N_calls)
```

```python
#Q2
from scipy.stats import binom

def question_2(N):
    p = 0.25
    prob_20_surgery = binom.pmf(20, N, p)
    prob_at_least_30_surgery = 1 - binom.cdf(29, N, p)
    expected_surgery = N * p

    print(f"i) Probability that exactly 20 patients need surgery: {prob_20_surgery:.4f}")
    print(f"ii) Probability that at least 30 patients need surgery: {prob_at_least_30_surgery:.4f}")
    print(f"iii) Expected number of patients requiring surgery: {expected_surgery:.2f}")

N_patients = 100
question_2(N_patients)
```
